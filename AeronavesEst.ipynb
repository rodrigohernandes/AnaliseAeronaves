> ### Aircraft_Phase (Fase da aeronave)
> ***

> ***
Pessoa ou empresa que responde pelo uso da aeronave.

### Tratando os dados

```python
#eliminar a coluna arit
#retirar duplicados/linhas vazias
dataset = dataset.drop('Arit',axis=1)
dataset = dataset.drop_duplicates()
dataset.dropna(how='all', inplace=True)
dataset['Ground_Casualties'].fillna(0, inplace=True)
dataset['Collision_Casualties'].fillna(0, inplace=True)

# Separando a coluna 'Incident Date' em dias, mes, ano para maior facilidade posteriormente
dataset['Dia_acidente'] = dataset['Incident_Date'].str[:2]
dataset['mes_acidente'] = dataset['Incident_Date'].str[3:6]
dataset['ano_acidente'] = dataset['Incident_Date'].str[-4:]
dataset.head()
```

```python
# alterando o dtype
dataset.Aircaft_Nature = dataset.Aircaft_Nature.astype('category')
dataset.Incident_Category = dataset.Incident_Category.astype('category')
dataset.Aircaft_Damage_Type = dataset.Aircaft_Damage_Type.astype('category')
dataset.Aircraft_Phase = dataset.Aircraft_Phase.astype('category')
dataset.Aircaft_Operator = dataset.Aircaft_Operator.astype('category')
```

```python
# causas mais comuns de acidente 

frequencia_causas = dataset['Incident_Cause(es)'].value_counts(ascending=False)
dfcausas = pd.DataFrame({'Frequencia':frequencia_causas})
dfcausas.head(5)
```

```python
# Grafico de barras comparando fatalidades e frequencia de acidentes por causa

df = df_cff[:11]
df['Causas'] = df.index


figure, bar_esq = plt.subplots(figsize=(16,8))
bar_esq.bar(df['Causas'], df['Frequencia'], color='royalblue', width=-0.4, align='edge')
bar_esq.set_ylabel('Frequência',size=20)

plt.xticks(rotation=90,fontsize=12)
plt.title('Fatalidades X Frequências por causa de acidentes',size=20,color='yellow')

bar_dir = bar_esq.twinx()
bar_dir.bar(df['Causas'], df['Fatalidades'], color='blue', width=0.4, align='edge')
bar_dir.set_ylabel('Fatalidades',size = 20)

plt.show()
```

```python
# Grafico de barras comparando fatalidades e frequencia de acidentes por causa (plotando usando o proprio df)

df = df_cff[:11]
df['Causas'] = df.index

df.plot(x= 'Causas',kind='bar',title='Fatalidades X Frequencias por causa de acidentes',figsize=(16,8))
```

```python
# Acidentes e Fatalidades com o passar dos anos

dfac = dataset[['ano_acidente','Fatalities']].groupby(['ano_acidente']).sum()
dfac['Frequencia'] = dataset['ano_acidente'].value_counts()
dfac = dfac.reset_index()
dfac.rename(columns={'Fatalities': 'Fatalidades','ano_acidente':'Ano'},inplace=True)
dfac.sort_values('Ano')
```

```python
# Gráfico de dispersão com a frequencia de acidentes com o passar dos anos

plt.figure(figsize=(20,6))
gd = sns.scatterplot(x='Ano', y="Frequencia",data=dfac)
gd.set_title("Acidentes com o passar dos anos",size=20,color='yellow')
gd.set_ylabel("Frequência",size=20)
gd.set_xlabel("Ano",size=20)
plt.xticks(rotation=90,size=8)
plt.show()
```

```python
# Grafico de Linhas com Acidentes e fatalidades com o passar dos anos

plt.figure(figsize = (20,8))
f_line = sns.lineplot(x = 'Ano', y = 'Frequencia',data = dfac)
fatal_line = sns.lineplot(x = 'Ano', y = 'Fatalidades',color = 'red',data = dfac)
plt.xticks(rotation=90,fontsize=8)
plt.title('Acidentes e fatalidades com o passar dos anos',size=20,color='yellow')
plt.legend(labels=["Frequencia","Fatalidades"])
plt.xlabel("Ano",size=20)
plt.ylabel("Contagem",size=20)
```

# Estatística descritiva de fatalidades

```python
media = dataset['Fatalities'].mean()
print(f'\nMédia:',round(media))

mediana = dataset['Fatalities'].median()
print(f'\nMediana:',mediana)

moda = dataset['Fatalities'].mode()
print(f'\nModa:',moda)

desvioMAD = dataset['Fatalities'].mad()
print(f'\nDesvio Médio Absoluto:',round(desvioMAD))

desvio = dataset['Fatalities']
print(f'\nDesvio Padrão:',desvio)
```
